```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sys
sys.path.append("/Users/tizianocausin/Desktop/backUp20240609/summer2025/ponce_lab/exp_set/python_scripts/src")
from dim_redu_anns.utils import get_relevant_output_layers
```

```python
path2res = "/Users/tizianocausin/OneDrive - SISSA/data_repo/exp_set_res/silico"
models = ["alexnet", "vit_b_16", "resnet18"]
```

```python
maxpool = 0
for m in models:
    layers = get_relevant_output_layers(m)
    var_explained = []
    for l in layers:
        if maxpool==1:
            path2file = f"{path2res}/var_explained_pca_{m}_{l}_maxpool_1000_PCs.csv"
        else:
            path2file = f"{path2res}/var_explained_pca_{m}_{l}_1000_PCs.csv"
        var_explained.append(pd.read_csv(path2file).to_numpy())
    counter = 0
    fig, axs = plt.subplots(1, len(layers), figsize=(len(layers)*3, 3))
    plt.tight_layout(pad=2.0)
    for el in var_explained:
        cdf = np.cumsum(el)
        axs[counter].plot(cdf)
        #axs[counter].set_xlabel("# components", fontsize=6)
        axs[counter].set_ylabel("% var explained", fontsize=6)
        axs[counter].set_ylim([0,1])
        #axs[counter].set_aspect('equal')
        axs[counter].set_title(f"{m} {layers[counter]}", fontsize=6)
        counter+=1
```
