```python
import numpy as np
import joblib
import matplotlib.pyplot as plt
import os, yaml, sys
ENV = os.getenv("MY_ENV", "dev")
with open("../../config.yaml", "r") as f:
    config = yaml.safe_load(f)
paths = config[ENV]["paths"]
sys.path.append(paths["src_path"])
from sampling.utils import compute_prob 
from sampling.sampling_comparisons import kld_calc
from sparsity_in_silico.sparsity_CNN import rust_sparseness, response_prob
import math
```

```python
ran_feats_path = f"{paths["results_path"]}/imagenet_val_alexnet_classifier.2_ran_features.pkl"
```

```python
data = joblib.load(ran_feats_path)
```

```python
var = np.var(data, axis=0)
m = np.mean(data, axis=0)
```

```python
#plt.hist(var, density=True)
counts, bins = np.histogram(var, bins=30)
plt.bar(bins[:-1], counts / counts.sum(), width=np.diff(bins))
```

```python
N = 100
klist = []
approx = 4
#var = np.round(var, approx)
bin_width = 10**(-(approx))
p_ground = compute_prob(var, bin_width)
#p_ground = compute_prob(m, bin_width)
#units_to_sample = 200
start, stop, step = (30, 45000, 500)
for N in range(start, stop, step):
    random_samples = np.random.choice(data.shape[0], N, replace=False)
    var_rs = np.var(data[random_samples,:], axis=0)
    #var_rs = np.var(data[random_samples,:units_to_sample], axis=0)
    #mean_rs = np.mean(data[random_samples,:], axis=0)
    #p_rs = compute_prob(mean_rs, bin_width)
    p_rs = compute_prob(var_rs, bin_width)
    k = kld_calc(p_ground, p_rs)
    klist.append(k)
```

```python
plt.plot(range(start, stop, step),klist, marker='o')
```

```python
rust_gt = rust_sparseness(data.T)
resp_prob_gt = response_prob(data.T)
```

```python
bins = np.arange(0,1,.01)
plt.hist(rust_gt, bins= bins)
plt.title("ground truth rust sparseness")
plt.figure()
plt.hist(resp_prob_gt,bins= bins)
plt.title("ground truth firing probability")
```

```python
sparseness_list = []
prob_list = []
start, stop, step = 10, 2000, 100
for N in range(start, stop, step):
    random_samples = np.random.choice(data.shape[0], N, replace=False)
    data_rs = data[random_samples,:]
    rust_rs = rust_sparseness(data_rs.T)
    resp_prob_rs = response_prob(data_rs.T)
    sparseness_list.append(rust_rs)
    prob_list.append(resp_prob_rs)
```

```python

n_plots = len(prob_list)
n_cols = 5
n_rows = math.ceil(n_plots / n_cols)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 10))
axes = np.array(axes).flatten()
subsampling_steps = range(start, stop, step)
for i, ax in enumerate(axes):
    if i < n_plots:
        ax.hist(prob_list[i], bins=bins)
        ax.set_title(f"{subsampling_steps[i]} samples")
    else:
        ax.axis("off")  # Hide any unused subplots

plt.tight_layout()
plt.show()
```
